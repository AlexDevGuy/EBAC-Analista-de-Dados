```python
#Importando Bibliotecas necessárias
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Comando Define a URL do arquivo robots.txt do GitHub.
URL_ROBOTS_TXT_ROBOTS_TXT = "https://github.com/robots.txt"

#código faz um request para a URL do robots.txt e armazena a resposta na variável resposta.
resposta = requests.get(URL_ROBOTS_TXT)

# código Extrai o conteúdo do robots.txt e armazena em resposta_data
resposta_data = resposta.text

#Codigo imprime o conteúdo robots.txt
print(resposta_data)
```

```python
#código Faz um request para a URL da página de Trending Repositories do GitHub e armazena a resposta na variável pagina_git_request.
pagina_git_request = requests.get('https://github.com/trending')

#código Extrai o conteúdo HTML da resposta (página de Trending Repositories) e armazena na variável pagina_github_html.
pagina_github_html = pagina_git_request.text
```

```python
#código Cria um objeto BeautifulSoup a partir do conteúdo HTML, utilizando o parser html.parser.
#O BeautifulSoup permite navegar e extrair informações da estrutura HTML da página.
soup = BeautifulSoup(pagina_github_html, 'html.parser')

# imprime o objeto BeautifulSoup
soup
```

```python
#Encontrando todos os projetos em destaque
#Procura e armazena em all_projects todos os elementos HTML da tag <article> que possuem a classe Box-row.
#Cada elemento representa um projeto em destaque.
all_projects = soup.find_all('article', attrs={'class':'Box-row'})
```

```python
#código Cria uma lista vazia github_list para armazenar informações extraídas de cada projeto.
github_list = []

#Extração de informações de cada projeto (linhas 15-42):

#Este bloco itera por cada projeto encontrado (all_projects)
#e extrai informações como nome, total de estrelas, estrelas hoje, forks e linguagens usadas.
for project in all_projects:
  #Nomes
  name_project = project.find('h2', attrs={'class':'h3 lh-condensed'})
  name_project = name_project.text
  names_github_project = name_project.split('\n')
  clean_name = [' '.join(i.split()) for i in names_github_project]
  name = '\n'.join(clean_name)
  name_project_final = name.replace('\n', ' ').strip()

  #total de estrelas
  all_stars = project.find('a', attrs={'class': 'Link Link--muted d-inline-block mr-3'})
  all_stars = all_stars.text
  all_stars = all_stars.replace('\n', ' ').strip()
  all_stars = int(all_stars.replace(',', ''))

  #Estrelas Hoje
  stars_today = project.find('span', attrs={'class': 'd-inline-block float-sm-right'})
  stars_today = stars_today.text
  stars_today = stars_today.replace('\n', ' ').strip()
  stars_today = int(stars_today.strip().split(' ')[0].replace(',',''))

  #Forks
  forks = project.find_all('a', attrs={'class': 'Link Link--muted d-inline-block mr-3'})[1]
  forks = forks.text
  forks = forks.replace('\n', ' ').strip()
  forks = int(forks.replace(',',''))

  #languages

  languages = project.find('span', attrs={'class': 'd-inline-block ml-0 mr-3'})
  if languages:
    languages = languages.text
    languages = languages.replace('\n', ' ').strip()


  github_list.append([name_project_final, languages, all_stars, stars_today, forks])
```

```python
#código Cria um DataFrame (github_df) a partir da lista github_list.
github_df = pd.DataFrame(github_list, columns=['Projects','Languages', 'all_stars', 'stars_today', 'forks'])

#Codigo Calcula o tamanho (len) do DataFrame e armazena em rank_len.
rank_len = len(github_df)

#codigo Cria uma nova coluna chamada Rank e preenche com valores do range 1 até o tamanho do DataFrame + 1
 #(para incluir todos os projetos classificados).
github_df['Rank'] = range(1, rank_len + 1)

#codigo Imprime o DataFrame (github_df) na tela.
github_df
```

```python
#codigo imprime Informações do DataFrame
#Exibe informações sobre o DataFrame, como o tipo de dados de cada coluna e o número de linhas e colunas.
github_df.info()
```

```python
#Exportando DataFrame para CSV
#Exporta o DataFrame para um arquivo CSV chamado github.csv.
#O parâmetro index=False evita que o índice seja salvo no CSV.
#O parâmetro sep=';' define o ponto e vírgula (;) como separador de campo no CSV.
github_df.to_csv('github.csv', index=False, sep=';')
```
