# **1. Introdução**

## 1.1 Objetivo

Os cartões de crédito são essenciais no sistema financeiro moderno. Este projeto de análise exploratória de dados (EDA) examinará os dados dos clientes de cartão de crédito para identificar tendências e padrões. O objetivo é extrair insights valiosos para otimizar negócios, aumentar a fidelidade do cliente, a receita e o desempenho geral.

## 1.2 Descrição dos dados

A fonte de dados da nossa análise é um arquivo de valores separados por vírgula (CSV) com dados fictícios para simular informações de contas bancárias criado por [André Perez](https://www.linkedin.com/in/andremarcosperez/). Você pode acessar este arquivo (credito.csv) neste [link](https://github.com/andre-marcos-perez/ebac-course-utils/tree/main/dataset).

## 1.3 Dados das colunas

* **idade** = idade do cliente
* **sexo** = gênero do cliente (F ou M)
* **dependentes** = número de dependentes do cliente
* **escolaridade** = nível de escolaridade do cliente
* **estado_civil** = estado civil do cliente
* **salario_anual** = faixa salarial anual do cliente
* **tipo_cartao** = tipo de cartão do cliente
* **qtd_produtos** = número de produtos adquiridos nos últimos 12 meses
* **iteracoes_12m** = número de interações/transações nos últimos 12 meses
* **meses_inativo_12m** = número de meses que o cliente ficou inativo
* **limite_credito** = limite de crédito do cliente
* **valor_transacoes_12m** = valor das transações dos últimos 12 meses
* **qtd_transacoes_12m** = número de transações nos últimos 12 meses

## 1.4 Ferramentas

Para este trabalho de análise de dados, utilizaremos três ferramentas: Amazon Web Services (AWS), S3 e Athena. O AWS é a plataforma de serviços em nuvem. O S3 é um serviço de armazenamento, e o Athena é um serviço de consulta que usa SQL para tratar os dados.

# **3. Conclusão**

Com base nos resultados obtidos neste projeto, os insights e as ações recomendadas são fornecidos a seguir.

## 3.1 Insights

1.   **Maior salário, maior limite:** Conforme mostrado, a única relação causal com o salário anual é o limite de crédito. A média dos produtos consumidos, os meses inativos e as iterações são praticamente iguais para todas as faixas salariais.

2.   **Precisa de mais clientes Premium:** Os dados revelam uma distribuição clara dos tipos de usuários, com a maioria caindo na categoria cartão azul, representando o tipo de cliente mais comum e padrão. Em contrapartida, os titulares de cartões Platinum, que representam nossos clientes mais premium, compõem apenas 0,08% do total de usuários. Talvez devêssemos investir mais para atrair clientes da categoria premium.

3.   **Enorme diferença nas transações entre os tipos de clientes:** Os valores médios das transações nos dizem que quanto mais premium o cartão, mais o cliente tende a gastar. Porém, como vimos anteriormente, há uma diferença significativa no número de clientes de cada tipo de cartão. Na realidade, a soma dos valores totais gastos em cada tipo de cartão apresenta uma diferença enorme que varia de 3 mil no nosso mais cartão premium (com poucos clientes) para mais de 4 milhões em nosso cartão inicial (com muitos clientes).

4.   **Baixa utilização de crédito no Premium:** O segmento premium tem limite de crédito médio de 17 mil, mas gasta apenas metade, enquanto o segmento standard tende a utilizar praticamente todo o seu limite.



## 3.2 Possíveis ações

1.   **Atrair clientes Premium:** Explore o potencial para atrair mais clientes premium, dado que a maioria dos clientes possui cartões azuis. Invista em marketing e serviços adaptados a este grupo para aumentar o número de titulares dos cartões Platinum e Gold.

2.   **Aumente os gastos premium:** Desenvolva planos e promoções destinadas a incentivar maiores gastos entre seu segmento de clientes premium, que tendem a não usar muito o limite disponível de seus cartões.
